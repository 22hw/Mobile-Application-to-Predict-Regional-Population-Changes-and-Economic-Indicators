```python
import numpy as np
import torch
from torch import nn
import pandas as pd
import matplotlib.pyplot as plt

# 모델 정의
class GRUModelWithDropout(nn.Module):
    def __init__(self, input_size, hidden_size, dropout_prob):
        super().__init__()
        self.rnn = nn.GRU(input_size, hidden_size, batch_first=True)
        self.dropout = nn.Dropout(p=dropout_prob)
        self.fc = nn.Linear(hidden_size, 1)
        self.input_size = input_size
        self.hidden_size = hidden_size

    def forward(self, x):
        x = x.view(-1, 1, self.input_size)
        h0 = torch.zeros(1, x.size(0), self.hidden_size)
        out, _ = self.rnn(x, h0)
        out = self.dropout(out)  # Dropout 적용
        out = self.fc(out[:, -1, :])
        return out

# 데이터 불러오기
df = pd.read_csv("/content/kangwon.csv")
x = df["the number of people"].to_numpy()
y = df["Gross regional product per capita"].to_numpy()
x = (x - x.mean()) / x.std()
y = (y - y.mean()) / y.std()

# 학습 데이터와 검증 데이터로 나누기
split_point = int(len(y) * 0.8)
x_train, y_train = x[:split_point], y[:split_point]
x_val, y_val = x[split_point:], y[split_point:]

# 모델 생성 (Dropout 적용)
input_size = 1
hidden_size = 128
dropout_prob = 0.2  # Dropout 확률 설정
model = GRUModelWithDropout(input_size=input_size, hidden_size=hidden_size, dropout_prob=dropout_prob)

# 데이터를 PyTorch 텐서로 변환
x_train = torch.tensor(x_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
x_val = torch.tensor(x_val, dtype=torch.float32)
y_val = torch.tensor(y_val, dtype=torch.float32).view(-1, 1)

# 미니 배치 학습을 위한 DataLoader 생성
batch_size = 32
train_dataset = torch.utils.data.TensorDataset(x_train, y_train)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# 모델 학습
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# 조기 종료 설정
best_val_loss = float('inf')
patience = 2000
counter = 0

for epoch in range(10000):
    for batch_x, batch_y in train_loader:
        y_pred = model(batch_x)
        loss = criterion(y_pred, batch_y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # 검증 데이터 손실 계산
    with torch.no_grad():
        y_val_pred = model(x_val)
        val_loss = criterion(y_val_pred, y_val)

    # 로그 출력
    if epoch % 1000 == 0:
        print(f"epoch {epoch}: loss {loss.item():.4f}, val_loss {val_loss.item():.4f}")


# 테스트 데이터 처리
x_test = x_val.clone().detach()
y_test = y_val.clone().detach().view(-1, 1)

# 테스트 데이터 예측
y_pred = model(x_test)

# 예측 결과 평가
mse = nn.MSELoss()(y_pred, y_test).item()
rmse = np.sqrt(mse)
print(f"RMSE: {rmse:.4f}")


```

```python
import numpy as np
import torch
from torch import nn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 데이터 불러오기
df = pd.read_csv("/content/kangwon.csv")

# 인구수와 1인당 지역내 총생산 추출
x = df["the number of people"].values
y = df["Gross regional product per capita"].values

# 데이터 전처리: 표준화
scaler_x = StandardScaler()
scaler_y = StandardScaler()
x = scaler_x.fit_transform(x.reshape(-1, 1))
y = scaler_y.fit_transform(y.reshape(-1, 1))

# 데이터 분할: 학습 데이터와 테스트 데이터
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 모델 정의
class MLPModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLPModel, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# 하이퍼파라미터 정의
input_size = 1
hidden_size = 64
output_size = 1
learning_rate = 0.001
num_epochs = 1000

# 모델 생성
model = MLPModel(input_size, hidden_size, output_size)

# 손실 함수 및 옵티마이저
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 학습 루프
for epoch in range(num_epochs):
    inputs = torch.tensor(x_train, dtype=torch.float32)
    targets = torch.tensor(y_train, dtype=torch.float32)

    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# 테스트 데이터에 모델 적용
model.eval()
with torch.no_grad():
    test_inputs = torch.tensor(x_test, dtype=torch.float32)
    predicted = model(test_inputs)

# 예측 결과 역 스케일링
predicted = scaler_y.inverse_transform(predicted.numpy())
y_test = scaler_y.inverse_transform(y_test)

# 예측 결과 평가 (RMSE)
rmse = np.sqrt(np.mean((predicted - y_test) ** 2))
print(f'Root Mean Square Error (RMSE): {rmse:.4f}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# 데이터 불러오기
df = pd.read_csv("/content/kangwon.csv")
year = df["Year"]  # Year 열을 따로 추출

# 데이터 전처리
y = df["Gross regional product per capita"].astype('float64')

# SARIMA 모델 생성
model = SARIMAX(y, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # 예시로 order와 seasonal_order를 지정
results = model.fit()

# SARIMA 모델의 예측
forecast = results.get_forecast(steps=len(y))
mean_forecast = forecast.predicted_mean
pret = mean_forecast.values.mean() - y.values.mean()
mean_forecast1 = mean_forecast - pret

# 다음 연도 예측 결과 추가
next_year = year.max() + 1
next_year_forecast = results.forecast(steps=1)  # 다음 연도의 예측값
next_year = year.append(pd.Series(next_year))
combined_forecast = pd.concat([mean_forecast1, next_year_forecast])

```

```python
# 그래프 그리기
print(next_year_forecast)
plt.figure(figsize=(12, 6))
plt.plot(year, y, label="실제 데이터", color='black')

plt.plot(next_year, combined_forecast, label="다음 년도 예측", color='blue')

plt.xlabel("연도")
plt.ylabel("1인당 지역내 총생산")
plt.xticks(np.arange(year.min(), year.max()+2, 1), rotation=45)
plt.yticks(np.arange(0, y.max() + 10000, 2500))
plt.legend()
plt.show()
```
